```python
''' Created a dictionary to store the information of the products by separating them into the categories such as 
Phones, Tablets, SIM cards etc. within the same structure: 
'''

PRODUCTS = {
    "Phones": [
        {"ItemCode": "BPCM", "Description": "Compact", "Price": 29.99},
        {"ItemCode": "BPSH", "Description": "Clam shell", "Price": 49.99},
        {"ItemCode": "RPSS", "Description": "Robo phone - 5inch 64GB memory", "Price": 199.99},
        {"ItemCode": "RPLL", "Description": "Robo phone - 6inch 256GB memory", "Price": 499.99},
        {"ItemCode": "YPLS", "Description": "Y-phone standard 6 inch 64GB memory", "Price": 549.99},
        {"ItemCode": "YPLL", "Description": "Y-phone deluxe 6 inch 256GB memory", "Price": 649.99}
    ],
    "Tablets": [
        {"ItemCode": "RTMS", "Description": "RoboTab - 8 inch screen 64GB memory", "Price": 149.99},
        {"ItemCode": "RTML", "Description": "RoboTab - 10 inch screen 128GB memory", "Price": 299.99},
        {"ItemCode": "YTLM", "Description": "Y-tab standard - 10 inch screen 128GB memory", "Price": 499.99},
        {"ItemCode": "YTLL", "Description": "Y-tab deluxe - 10 inch screen 256GB memory", "Price": 599.99}
    ],
    "SIM cards": [
        {"ItemCode": "SMNO", "Description": "Sim free (no SIM card)", "Price": 0.00},
        {"ItemCode": "SMPG", "Description": "Pay as you go (with SIM card)", "Price": 9.99}
    ],
    "Cases": [
        {"ItemCode": "CSST", "Description": "Standard", "Price": 0.00},
        {"ItemCode": "CSLX", "Description": "Luxury", "Price": 50.00}
    ],
    "Chargers": [
        {"ItemCode": "CGCR", "Description": "Car", "Price": 19.99},
        {"ItemCode": "CGHM", "Description": "Home", "Price": 15.99}
    ]
}
```

```python

def validate_item_code(new_item_code):
    #function for validating item codes which returns True if the item code exists, otherwise False.
    item_code = new_item_code.upper()
    for category,items in PRODUCTS.items():
        for item in items:
            if item["ItemCode"] == item_code:
                
                return True
    
    return False
```

```python
#Example usage:
validate_item_code("YTLM")
```

```python
def validated_new_data_entry(new_item_code):
    
    #function for adding validated new product entries by using another function validate_item_code()
    
    if validate_item_code(new_item_code) == False:
        description = input("Enter the product description: ")
        try:
            price = round(float(input("Enter the product price: ")), 2)
        except ValueError:
            print("Invalid price. (Please, do not leave blank or don't use symbols except dot (.) for decimal point.)")
        category = str.capitalize(input("Enter the product category (Phones, Tablets, SIM cards, Cases, Chargers): "))
        if category in PRODUCTS:
            PRODUCTS[category].append({"ItemCode": new_item_code, "Description": description, "Price": price})
            print(f"New product added to {category} category.")
            print(f"Item code {new_item_code} is now valid.")
        else:
            print("Category is invalid. \nProduct not added.")
    else:
        print("There is an existing item code as", new_item_code,". Therefore, product not added.")
```

```python
#Example usage:

validated_new_data_entry("XYZ1")
```

```python
#checking the total number of items in the PRODUCTS dictionary after adding or deleting items.

def count_total_items():
    total = 0
    for category, items in PRODUCTS.items():
        for item in items:
            total += 1
    print("Total numbers of items:", total)

count_total_items()
```

```python
#Deleting item
def delete_item_by_code():
    #function to delete an item with the item code of itself.
    
    while True:
        item_code = input("Enter valid item code(s) to delete (or pres any key except valid item code to EXIT): ")
        
        if validate_item_code(item_code):
            found = False

            for category, items in PRODUCTS.items():
                for item in items[:]:
                    if item["ItemCode"] == item_code:
                        items.remove(item)
                        print(f"Item with code {item_code} has been deleted from {category} category.")
                        found = True
                        break
                if found:
                    break
                
        else:
            print(f"Item with code {item_code} does not exist. Deletion cancalled.")
            break
```

```python
#Example usage:
delete_item_by_code()
```
